Exemplo de DataSet que usa KNN para fazer predição de spam com acurácia de 91%
referência : https://www.kaggle.com/datasets/uciml/sms-spam-collection-dataset

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/spam.csv', delimiter=',',encoding='latin-1')
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],axis=1,inplace=True)
df.sample(10) #10 linhas aleatorias
```

```python
#tratamento dos dados
df["v1"] = (df["v1"] == "spam").astype(int)

x, y=df['v2'].values,df['v1'].values

x = x.tolist()
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(x)
```

```python
#separando em 80%treino e 20%teste
train_x,test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
#fazendo o treinamento do modelo
knn_model = KNeighborsClassifier()
knn_model.fit(train_x, train_y)

test_pre = knn_model.predict(test_x)
```

```python
#resultados do modelo
print("Acurácia:", accuracy_score(test_y, test_pre))
print("Relatório de Classificação:\n", classification_report(test_y, test_pre))
```

```python
# Selecionar 10 mensagens aleatórias do conjunto de teste
num_samples = 10
indices = np.random.choice(test_x.shape[0], num_samples, replace=False)

# Obter as mensagens e os rótulos reais
sample_messages = test_x[indices]
sample_labels = test_y[indices]

# Fazer a predição usando o modelo treinado
predicted_labels = knn_model.predict(sample_messages)

# Exibir os resultados
print("\nValidação Manual de Classificação de Spam:")
for i in range(num_samples):
    original_message = vectorizer.inverse_transform(sample_messages[i])[0]  # Reconstruir o texto original
    predicted_label = "Spam" if predicted_labels[i] == 1 else "Não Spam"
    actual_label = "Spam" if sample_labels[i] == 1 else "Não Spam"

    print(f"\nMensagem: {' '.join(original_message)}")
    print(f"Rótulo Real: {actual_label}")
    print(f"Predição do Modelo: {predicted_label}")
    print("-" * 80)
```
