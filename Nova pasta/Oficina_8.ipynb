Exemplo de PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_wine
import pandas as pd

wine = load_wine()

# Converter para DataFrame do Pandas
df_wine = pd.DataFrame(wine.data, columns=wine.feature_names)

# Adicionar a coluna de rótulo (classe do vinho)
df_wine["target"] = wine.target

df_wine.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_wine
import pandas as pd

samples = load_wine()
target = samples.target
samples = samples.data
scaler = StandardScaler()
scaled_samples = scaler.fit_transform(samples)


# Create a PCA model with  components: pca
pca = PCA(n_components=2)

# Fit the PCA instance to the scaled samples
pca.fit(scaled_samples)

# Transform the scaled samples: pca_features
pca_features = pca.transform(scaled_samples)

# Print the shape of pca_features
#print(pca_features)
pca_features.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

knn = KNeighborsClassifier(n_neighbors=6)

X_train, X_test, y_train, y_test = train_test_split(pca_features, target, test_size=0.3, random_state=42)

knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Criando um DataFrame de exemplo (1 = produto comprado, 0 = produto não comprado)
data = {
    'Leite': [1, 1, 0, 1, 0, 1],
    'Pão': [1, 1, 1, 1, 0, 1],
    'Café': [0, 1, 1, 0, 1, 1],
    'Ovos': [1, 0, 0, 1, 1, 0],
    'Manteiga': [0, 1, 1, 1, 1, 1]
}

df = pd.DataFrame(data)

# Exibir o dataset de transações
df.head()
```

```python
# Aplicar Apriori para encontrar os conjuntos de itens frequentes
frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)

# Exibir os conjuntos frequentes
frequent_itemsets
```

```python
# Gerar regras de associação
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Exibir as regras geradas
rules
```

- Quando compram leite tem 83% de chance de comprar pão
- Quando compram café tem 83% de chance de comprar manteiga
