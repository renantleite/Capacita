```python
import pandas as pd
import numpy as np
```

```python
#criação do DataFrame com os dados que foram mencionados na oficina.
dados = {

    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo'],

    'Idade': [28, 34, 29, None, 42],

    'Cidade': ['São Paulo', 'Rio de Janeiro', None, 'Curitiba', 'Belo Horizonte'],

    'Vendas': [150, 200, 300, 400, None]

}
df = pd.DataFrame(data = dados)
```

```python
# Mostrando o DataFrame
df
```

```python
# Filtrando apenas os valores onde a idade seja maior que 30
df.where(df["Idade"] > 30).dropna()
```

```python
# Calculando a media das idades dos clientes
media_idade = np.mean(df["Idade"])
media_idade
```

```python
#trocando valores Nulos por Desconhecido na coluna Cidade
df.loc[df["Cidade"].isna(), "Cidade"] = "Desconhecido"
```

```python
df
```

```python
#Calculando a soma das vendas para saber o total
Total_vendas = np.sum(df["Vendas"])
Total_vendas
```
